```python
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.preprocessing import StandardScaler, MinMaxScaler, RobustScaler, Normalizer, Binarizer
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
import pandas as pd

digits = load_digits()
X, y = digits.data, digits.target

digits

#датафрейм

```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)


#разделение на обуч и тест выборки
```

```python
classifiers = {
    "Logistic Regression": LogisticRegression(max_iter=10000, solver='saga'),
    "KNN": KNeighborsClassifier(),
    "Random Forest": RandomForestClassifier(),
    "SVM": SVC()
}  #классификаторы


scalers = {
    "StandardScaler": StandardScaler(),
    "MinMaxScaler": MinMaxScaler(),
    "RobustScaler": RobustScaler(),
    "Normalizer": Normalizer(),
    "Binarizer": Binarizer()
}   #методы преобразования

```

```python
results = {}

for scaler_name, scaler in scalers.items():
    X_train_scaled = scaler.fit_transform(X_train)

    for clf_name, clf in classifiers.items():
        scores = cross_val_score(clf, X_train_scaled, y_train, cv=5, scoring='accuracy') # через кросс-валидацию
        avg_score = scores.mean()

        if scaler_name not in results:
            results[scaler_name] = {}
        results[scaler_name][clf_name] = avg_score

df_results = pd.DataFrame(results).T
print(df_results)
# результат в датасетик
```
